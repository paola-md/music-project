```python
import os
from midi2audio import FluidSynth
```

```python
midi_files = []
for path, subdirs, files in os.walk('../data/'):
    for name in files:
        midi_files.append(os.path.join(path, name))
midi_files = [f for f in midi_files if '.midi' in f]
```

```python
midi_files
```

```python
for input_file in midi_files:
    fs = FluidSynth('../data/sound-fonts/Essential Keys-sforzando-v9.6.sf2')
    output_file = input_file.replace('/', '-').replace('-data-maestro-v3.0.0-', '/data/wav_files/').replace('.midi', '.wav')
    fs.midi_to_audio(input_file, output_file)
```
